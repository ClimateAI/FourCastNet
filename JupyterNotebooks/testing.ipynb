```python
import s3fs
import h5py
import numpy as np
import matplotlib.pyplot as plt
s3 = s3fs.S3FileSystem()
```

```python
path = 's3://climate-ai-data-science-sandbox-fourcastnet/data/out_of_sample/2018.h5'
f = h5py.File(s3.open(path, 'rb'), 'r')
f.keys()
```

```python
small = f['small']
```

```python
d = f['fields']
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
im = plt.imshow(d[0,20,:,:])
plt.colorbar(im)
plt.show()
```

```python
d[0, 20, :, :]
```

```python

# Create a figure with 21 subplots
fig, axes = plt.subplots(nrows=7, ncols=3, figsize=(10, 20))

# Flatten the axes array to facilitate iterating over the subplots
axes = axes.flatten()

for c in range(21):
    # Plot the data for each 'c' value
    im = axes[c].imshow(d[0, c, :, :])
    axes[c].set_title(f'c = {c}')  # Set the title for each subplot
    fig.colorbar(im, ax=axes[c])

# Hide any remaining empty subplots
for c in range(21, len(axes)):
    axes[c].axis('off')

# Adjust spacing between subplots
plt.tight_layout()

plt.show()
```
